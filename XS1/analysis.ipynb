```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
# import csv
from matplotlib import rc, rcParams
```

```python
with open('uranium.spe', mode ='r') as file:    
    lines = file.readlines()

lines = [line.strip() for line in lines]
data = [int(line) for line in lines if line.isdigit()]
print(data)

```

```python
plt.hist(data, 50)
plt.title("Uranium counts", fontsize=20)
plt.xlabel(r'Counts', fontsize=20, labelpad=20)
plt.ylabel(r'Number of bins', fontsize=20)
plt.show()
```
