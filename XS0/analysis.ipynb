```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
from scipy.special import factorial
import math
# import csv
from matplotlib import rc, rcParams
```

```python
data = [9.45711821922, 9.43696722061, 9.65821724749, 9.53421335027]
statlenerrors = np.array([0.5 / 37.4, 0.5 / 37.4, 0.5 / 56.3, 0.5 / 56.3])
stattimeerrors = np.array([0.035 / 12.495 * 1/math.sqrt(10), 0.0545 / 37.525 * 1/math.sqrt(30), 0.03 / 15.17 * 1/math.sqrt(10), 0.055 / 45.805 * 1/math.sqrt(30)])
staterrors = (statlenerrors + stattimeerrors * 2) * data
print(staterrors)
print(f"avg: {sum(data)/len(data)}")
print(f"best avg: {sum([dat / staterrors[i] ** 2 for i, dat in enumerate(data)]) / np.sum(1 / staterrors ** 2)}")
print(math.sqrt(np.sum(staterrors ** 2)) / 4)
print(f"best avg error: {4 / np.sum(1 / staterrors ** 2)}")

plt.errorbar(np.arange(1, len(data)+1), data, yerr = staterrors, elinewidth=0.6, linestyle="dotted")
plt.plot(np.arange(1, len(data)+1), [sum(data)/len(data)]*4, label="average, error +/- 0.0587")
plt.plot(np.arange(1, len(data)+1), [9.565]*4, label="goes through all statistical error ranges")

plt.xlabel(r'Trial', fontsize=20, labelpad=20)
plt.ylabel(r'Obtained $g$', fontsize=20)
plt.legend(loc="lower right",fontsize=8, ncol = 1)
plt.xticks([1,2,3,4])
plt.show()
```
